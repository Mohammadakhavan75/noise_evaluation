```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
from glob import glob
```

```python
cifar10_in_shifts_out_binary = {}
with open("cifar10_in_shifts_out_binary.log", "r") as f:
    for line in f:
        if line == '\n':
            continue
        else:
            (key, val) = line.split(':')
            cifar10_in_shifts_out_binary[key] = float(val)

cifar10_in_shifts_out_scone = {}
with open("cifar10_in_shifts_out_scone.log", "r") as f:
    for line in f:
        if line == '\n':
            continue
        else:
            (key, val) = line.split(':')
            cifar10_in_shifts_out_scone[key] = float(val)

noise_evaluation_report = {}
with open("noise_evaluation_report.log", "r") as f:
    for line in f:
        if line == '\n':
            continue
        else:
            (key, val) = line.split(':')
            noise_evaluation_report[key] = float(val)
```

```python
plt.scatter(list(cifar10_in_shifts_out_scone.keys()),list(cifar10_in_shifts_out_scone.values()))
plt.scatter(list(cifar10_in_shifts_out_binary.keys()),list(cifar10_in_shifts_out_binary.values()))
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.scatter(list(cifar10_in_shifts_out_scone.keys()),list(cifar10_in_shifts_out_scone.values()), label='scone')
ax.scatter(list(cifar10_in_shifts_out_binary.keys()),list(cifar10_in_shifts_out_binary.values()), label='binary')

plt.xticks(rotation=90, fontsize='xx-small')
plt.legend()
plt.ylabel("ROC-AUC")
plt.savefig('scone_vs_binary.jpg', dpi=1000)
plt.show()
```
